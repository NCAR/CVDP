# End-to-End Workflow Example

```python
from cvdp.io import get_input_data
from cvdp.diag import compute_seasonal_avgs, compute_seasonal_stds
from cvdp.vis import plot_seasonal_means, CVDPNotebook
```

## I/O Step

```python
ref_datasets, sim_datasets = get_input_data("example_config.yaml")
```

## Computation Step

```python
seasonal_avgs = compute_seasonal_avgs(sim_datasets["LENS1_TREFHTMX"])
ensemble_avgs = seasonal_avgs.mean(dim="member").compute()
```

## Visualization Step

```python
seasonal_ensemble_fig = plot_seasonal_means(ensemble_avgs)

cvdp_nb = CVDPNotebook()
cvdp_nb.add_figure_cell(seasonal_ensemble_fig, section_name="Seasonal Ensemble Means")
cvdp_nb.save_notebook()
```
